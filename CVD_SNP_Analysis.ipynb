```python
# Matrix Manipulation Packages
import pandas as pd
import numpy as np

# Statistical Analysis Packages
from scipy.spatial.distance import jensenshannon
from scipy.stats import chi2_contingency

# Visualization Packages
import matplotlib.pyplot as plt
```

```python
# Load HF Mutations Statistics (61 Patient Cohort)
sumHF_CVD = pd.read_csv('SNP_61_Analysis_Materials/Summaries/Summary_Mutation_Analysis_HF.csv')
sumHF_CVD.set_index('Gene', inplace=True)

# Total Mutations for Genes
gene_mutations = sumHF_CVD.sum(axis = 0)[1:]

# Generate Proportions for Genes
gene_proportions = sumHF_CVD.iloc[:,1:].div(sumHF_CVD.sum(axis=1), axis=0)

# Generate Proportion Distrubution for HF Genes  
average_proportion = np.mean(gene_proportions, axis = 0)

# Jensen-Shannon Divergence
js_divergences = gene_proportions.apply(lambda x: jensenshannon(x, average_proportion), axis = 1)
divergence_df = pd.DataFrame({'Gene': gene_proportions.index, 'Jensen-Shannon Divergence': js_divergences})
unique_genes = divergence_df.sort_values(by='Jensen-Shannon Divergence', ascending=False)
unique_genes.to_csv('SNP_61_Analysis_Materials/Jensen-Shannon/Jensen-Shannon-Divergence_scoresHF.tsv', index = None, sep = '\t')
```
